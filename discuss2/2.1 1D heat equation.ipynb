```julia
using DifferentialEquations
using Plots
```

## Heat Equation

$$
\frac{\partial u}{\partial t} = \alpha\frac{\partial^2 u}{\partial x^2}
$$

```julia
Δx = 0.01
x = Δx:Δx:1-Δx

N = length(x)
A = zeros(N,N)
for i in 1:N, j in 1:N
  abs(i-j)<=1 && (A[i,j]+=1)
  i==j && (A[i,j]-=3)
end
A = A/(Δx^2)

ts = (0.0,4.0)
@. U0 = sin(2π*x)+2cos(3π*x)+4sin(0.3π*x)
boundary = zeros(N); boundary[1] = U0[1]/(Δx^2)
heat(u,p,t) = p*A*u + boundary;
```

```julia
prob = ODEProblem(heat,U0,ts,2.0)
sol = solve(prob);
```

```julia
plot(sol.u,label="")
```
